```python
from tensorflow import keras
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heart.csv')
df = df.sample(frac=1).reset_index(drop=True)
```

```python
y_train = df['target'].iloc[:int(df.shape[0] * 0.8)]
y_test = df['target'].iloc[int(df.shape[0]* 0.8) :]
```

```python
df.iloc[df['age'].idxmin(), 0]
df.iloc[df['age'].idxmax(), 0]

def AgeClassification (row):
    if row >= 70:
        return 1
    elif row >= 60:
        return 0.8
    elif row >= 50:
        return 0.6
    elif row >= 40:
        return 0.4
    elif row >= 30:
        return 0.2
    else:
        return 0
```

```python
x_train = df.iloc[:int(df.shape[0] * 0.8),:-1]
x_test = df.iloc[int(df.shape[0] * 0.8):,:-1]
x_train = np.array(x_train)
x_test = np.array(x_test)
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(13, input_shape=(13, ), use_bias = True, kernel_initializer='glorot_uniform', bias_initializer='zeros'))
model.add(keras.layers.Dense(13, activation = "relu", use_bias = True, kernel_initializer='glorot_uniform', bias_initializer='RandomNormal'))
model.add(keras.layers.Dense(8, activation = "relu", use_bias = True, kernel_initializer='glorot_uniform', bias_initializer='RandomNormal'))
model.add(keras.layers.Dense(8, activation = "relu", use_bias = True, kernel_initializer='glorot_uniform', bias_initializer='RandomNormal'))
model.add(keras.layers.Dense(8, activation = "relu", use_bias = True, kernel_initializer='glorot_uniform', bias_initializer='RandomNormal'))
model.add(keras.layers.Dense(8, activation = "relu", use_bias = True, kernel_initializer='glorot_uniform', bias_initializer='RandomNormal'))
model.add(keras.layers.Dense(4, activation = "relu", use_bias = True, kernel_initializer='glorot_uniform', bias_initializer='RandomNormal'))
model.add(keras.layers.Dense(4, activation = "relu", use_bias = True, kernel_initializer='glorot_uniform', bias_initializer='RandomNormal'))
model.add(keras.layers.Dense(1, activation = "sigmoid"))

opt = keras.optimizers.Adam(lr=0.001)
model.compile(optimizer = opt, loss = keras.losses.BinaryCrossentropy(), metrics = ['accuracy'])

model.summary()
```

```python
model.fit(x_train, y_train, batch_size = 10, epochs = 300)
```

```python
pred = model.predict(x_test)
pred = np.array(pred)
pred = np.round(pred)
```

```python
y_test = np.array(y_test)

plt.subplots(figsize = (15, 5))
plt.plot(y_test[:100], color='green')
plt.plot(pred[:100], color='red')
plt.title('Heart Disease')
plt.ylabel('Yes / No')

plt.show()
```

```python
def Calculation(p, t):
    count = 0
    for i in range(p.shape[0]):
        if(p[i] == t[i]):
            count+=1
    return count / p.shape[0]
print('Accuracy: ', Calculation(pred, y_test))
```
